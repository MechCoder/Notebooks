```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

rng = np.random.RandomState(0)
# Model the leaf of each tree with a gaussian.
# Leaf1 ~= P(Y | L=1) ~ N(2, 1) 
mean1 = 2
std1 = 1
leaf1 = rng.normal(mean1, std1, 10000)

# Leaf2 ~= P(Y | L=2) ~ N(2, 3) 
mean2 = 2
std2 = 3
leaf2 = rng.normal(mean2, std2, 10000)

# Leaf3 ~= P(Y| L=3) ~ N(3, 2) 
mean3 = 3
std3 = 2
leaf3 = rng.normal(mean3, std3, 10000)

numBins = 50

plt.hist(leaf1, numBins,color='green', alpha=0.8)
plt.show()

plt.hist(leaf2, numBins, color='green', alpha=0.8)
plt.show()

plt.hist(leaf3, numBins, color='green', alpha=0.8)
plt.show()
```

```python
# According to the paper we should end up with

(1.0 / 3) * (1 + 4 + 9) + (1.0 / 3) * (4 + 4 + 9) - ((2 + 3 + 2) / 3)**2 
```

```python
# Y = (Y1 + Y2 + Y3) / 3
# Equivalent to randomly sampling from each leaf and averaging to make prediction

P = (leaf1 + leaf2 + leaf3) / 3

plt.hist(P, numBins, color='green', alpha=0.8)
plt.show()


print(np.var(P))
```

```python
# Randomly pick a leaf and use the random sample from that leaf to make predictions

leaves = {0: leaf1, 1: leaf2, 2: leaf3}
leaf_choices = rng.choice([0, 1, 2], 10000)
samples = np.arange(10000)

random_samples = []
for i in range(10000):
    random_leaf = leaf_choices[i]
    random_samples.append(leaves[random_leaf][rng.choice(samples)])

plt.hist(random_samples, numBins, color='green', alpha=0.8)
plt.show()

print(np.var(random_samples))
```
